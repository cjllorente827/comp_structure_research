# Critical Density
Critical Density: average density at which the universe is closed

$$
\rho_c(z) = \frac{3 H^2(z)}{8\pi G}
$$

$$
H^2(z) = H_0^2 \left[ \Omega_r(1+z)^4 + \Omega_m(1+z)^3 + (1-\Omega_{tot})(1+z)^4 + \Omega_\Lambda \right]
$$

The virial radius is the radius at which the mean density is the critical density multiplied by an overdensity parameter
$$
\rho(r = r_{vir}) = \Delta_c \rho_c
$$

Usually $\Delta_c = 200$


```python
Ωm = 0.3
ΩΛ = 0.7
H0 = 70 * u.km /u.s /u.Mpc
H = lambda z : H0 * np.sqrt( Ωm*(1+z)**3 + ΩΛ )

z = 0.0
ρc = ( 3 * H(z)**2 /(8*np.pi*const.G) ).to('g/cm^3')
print(ρc)
```

For a standard cosmological model with $\Omega_m = 0.3, \Omega_\Lambda = 0.7, H_0 = 70$ km/s/Mpc, $\rho_c = 9.2 \times 10^{-30}$ g/cm$^3$

# Virial Mass and Radius
$$
M_{vir} = \frac{4}{3}\pi r_{vir}^3 \Delta_c \rho_c
$$

The Navarro-Frenk-White profile can be used as a model for the density distribution of a dark matter halo.
$$
\rho(r) = \frac{\delta_c\rho_c}{\displaystyle \frac{r}{R_s} \left(1+\frac{r}{R_s} \right)^2}
$$

# Redshift as a function of time (and vice versa)

From <i>The Cosmic Time in Terms of the Redshift</i> Carmeli, Hartnett, and Oliveira (2008)
$$
t(z) = \frac{2 / H_0}{1+(1+z)^2} \approx \frac{28 \text{ Gyr}}{1+(1+z)^2}
$$
for $H_0 = 70$ km/s/Mpc


```python
z = np.linspace(0, 15)
t = lambda z: 28/(1+(1+z)**2)

fig, ax = plt.subplots(2,2,figsize=(28,14))
ax[0,0].plot(t(z),z)
ax[0,0].set_xlabel("Time (Gyr)")
ax[0,0].set_ylabel("Redshift")

ax[1,0].plot(z,t(z))
ax[1,0].set_ylabel("Time (Gyr)")
ax[1,0].set_xlabel("Redshift")
ax[1,0].invert_xaxis()

z_zoom_in = np.linspace(0, 2)

ax[0,1].plot(t(z_zoom_in),z_zoom_in)
ax[0,1].set_xlabel("Time (Gyr)")
ax[0,1].set_ylabel("Redshift")

ax[1,1].plot(z_zoom_in,t(z_zoom_in))
ax[1,1].set_ylabel("Time (Gyr)")
ax[1,1].set_xlabel("Redshift")
ax[1,1].invert_xaxis()
```
